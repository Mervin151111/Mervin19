# 4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

```

# Count the number of guests for each hotel type
hotel_counts = df['hotel'].value_counts()

# Specify the degree of separation for each slice
explode = [0.1 if i == 0 else 0 for i in range(len(hotel_counts))]

# Create the pie chart with the explode parameter
plt.pie(hotel_counts, labels=hotel_counts.index, autopct='%.2f', explode=explode)
plt.title('Guests by Hotel Type')
plt.show()

# Remark of this analysis -> The takeaway from this analysis is that the majority of guests stayed at the City Hotel, 
# followed by the Resort Hotel, and there is a notable difference in the number of guests between the two hotel types.
```

```
# Count the number of canceled and not canceled bookings
canceled_count = df['is_canceled'].value_counts()[1]
not_canceled_count = df['is_canceled'].value_counts()[0]

# Print the results
print("Number of canceled bookings:", canceled_count)
print("Number of not canceled bookings:", not_canceled_count)

# Remark for this analysis -> # The takeaway from this analysis is that a significant number of bookings in the dataset were canceled, as
# indicated by the higher count of canceled bookings compared to not canceled bookings.This information can be useful in understanding booking
# patterns and identifying potential factors that contribute to cancellations, such as hotel policies,customer preferences, or external events.
```

```
# Count the number of cancellations and non-cancellations

cancel_counts = df['is_canceled'].value_counts()
not_canceled = cancel_counts[0]

if len(cancel_counts) > 1:
    canceled = cancel_counts[1]
else:
    canceled = 0

explode = (0, 0.1) # Set the slice separation for the second slice
plt.pie([not_canceled, canceled], labels=['Not Canceled', 'Canceled' if canceled > 0 else ''], autopct='%1.1f%%', explode=explode)
plt.title('Proportion of Canceled vs Not Canceled Bookings')
plt.show()

# Remarks for this analysis -> #The takeaway from this analysis is that a significant proportion of bookings in the dataset were canceled, 
# as indicated by the larger slice of the pie chart representing canceled bookings. The pie chart is an effective way to quickly convey 
# this information and provides a visual comparison between canceled and not canceled bookings.
```

```
# Count the number of repeated and non-repeated guests
guest_counts = df['is_repeated_guest'].value_counts()

# Calculate the percentage of repeated guests
total_guests = guest_counts.sum()
repeated_percent = guest_counts[1] / total_guests * 100
not_repeated_percent = guest_counts[0] / total_guests * 100

# Create the pie chart with separation of each slice
labels = ['Repeated Guests', 'Non-Repeated Guests']
sizes = [repeated_percent, not_repeated_percent]
explode = (0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title('Percentage of Repeated Guests')

plt.show()

# Reamrk for analysis -> # The takeaway from this analysis is that a relatively small percentage of guests in the dataset were repeated 
# guests, as indicated by the smaller slice of the pie chart representing repeated guests. This information can be useful in understanding 
# guest loyalty and identifying factors that contribute to guest retention, such as hotel amenities, customer service, or loyalty programs.
```

```
# Count the number of guests by country
guest_counts = df['country'].value_counts()

# Calculate the percentage of guests from the top 10 countries
total_guests = guest_counts.sum()
top_10_guests = guest_counts[:10].sum()
top_10_percent = top_10_guests / total_guests * 100
others_percent = 100 - top_10_percent

# Create the pie chart
labels = ['Top 10 Countries', 'Others']
sizes = [top_10_percent, others_percent]
explode = [0.1, 0]  # Explode the first slice by 10%
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title('Percentage of Guest Distribution by Country')

plt.show()

# Reamrk for this analysis -> # The takeaways from this analysis are that a relatively small number of countries account for a 
# significant proportion of the total guests in the dataset, as indicated by the larger slice of the pie chart representing the 
# top 10 countries. This information can be useful in understanding the geographic distribution of guests and identifying areas for 
# targeted marketing or outreach efforts.
```

```
# Combine day, month, and year information to create a complete date column
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'].astype(str))

# Create a new column for booking lead time in days
df['lead_time'] = (df['arrival_date'] - pd.to_datetime(df[df['is_canceled']==0]['arrival_date'])).dt.days

# Create a new column for ADR per stay
df['adr_per_stay'] = df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])

# Group the data by month and calculate the mean lead time and ADR per stay
monthly_data = df.groupby('arrival_date_month').agg({'lead_time': 'mean', 'adr_per_stay': 'mean'})

# Create a scatter plot of lead time vs ADR for each month
fig, ax = plt.subplots()
for month, data in monthly_data.groupby('arrival_date_month'):
    ax.scatter(data['lead_time'], data['adr_per_stay'], label=month)

ax.legend()
ax.set_xlabel('Lead Time (Days)')
ax.set_ylabel('ADR per Stay')
ax.set_title('Lead Time vs ADR by Month')
plt.show()

# Remarks of tha analysis -> This analysis shows the relationship between lead time (the time between booking and arrival) and 
# ADR (average daily rate) per stay for each month. The scatter plot shows a positive correlation between lead time and ADR, with 
# longer lead times generally associated with higher ADR per stay. The plot also shows some variation between months, with some months
# having a stronger correlation than others. This information can be useful for understanding booking trends and optimizing pricing 
# strategies for different months.
```

```
# Filter the dataset to include only non-canceled bookings
df = df[df['is_canceled'] == 0]

# Group the dataset by arrival_date_month and calculate the average number of bookings and the average price per night
avg_bookings = df.groupby('arrival_date_month')['hotel'].count().reset_index(name='avg_bookings')
avg_price = df.groupby('arrival_date_month')['adr'].mean().reset_index(name='avg_price')

# Merge the two dataframes on arrival_date_month
monthly_data = pd.merge(avg_bookings, avg_price, on='arrival_date_month')

# Create a bar chart to visualize the average number of bookings and average price per month
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()
monthly_data.plot(x='arrival_date_month', y='avg_bookings', ax=ax1, legend=False, kind='bar', color='C0', width=0.4)
monthly_data.plot(x='arrival_date_month', y='avg_price', ax=ax2, legend=False, color='C1', marker='o')
ax1.set_ylabel('Average Number of Bookings', fontsize=12)
ax2.set_ylabel('Average Price per Night ($)', fontsize=12)
ax1.set_xlabel('Month', fontsize=12)
plt.xticks(rotation=45)
plt.title('Average Monthly Bookings and Prices', fontsize=14)
plt.show()

# This code calculates average number of bookings and average price per night by month. This helps to identify which month
# has the lowest number of bookings or lowest prices

# Reamrk of the analysis -> The above analysis provides insights into the average monthly bookings and prices for non-canceled bookings. 
# The analysis shows that the highest number of bookings were made in August, followed by July and May. The lowest number of bookings 
# were made in January, February, and December. The average price per night is highest in August and lowest in January. The analysis 
# provides valuable insights into the hotel industry and can help hotel managers to plan their pricing and promotions.
```

```
# Filter the dataset to include only non-canceled bookings
df = df[df['is_canceled'] == 0]

# Group the dataset by hotel, arrival_date_month, and customer_type, and calculate the average number of bookings and the average price per night
grouped_data = df.groupby(['hotel', 'arrival_date_month', 'customer_type']).agg(avg_bookings=('hotel', 'count'), avg_price=('adr', 'mean')).reset_index()

# Create separate bar charts to visualize the average number of bookings and average price per night for each hotel and customer_type group
for hotel in df['hotel'].unique():
    for customer_type in df['customer_type'].unique():
        subset = grouped_data[(grouped_data['hotel'] == hotel) & (grouped_data['customer_type'] == customer_type)]
        if not subset.empty:
            fig, ax1 = plt.subplots(figsize=(12, 6))
            ax2 = ax1.twinx()
            subset.plot(x='arrival_date_month', y='avg_bookings', ax=ax1, legend=False, kind='bar', color='C0', width=0.4)
            subset.plot(x='arrival_date_month', y='avg_price', ax=ax2, legend=False, color='C1', marker='o')
            ax1.set_ylabel('Average Number of Bookings', fontsize=12)
            ax2.set_ylabel('Average Price per Night ($)', fontsize=12)
            ax1.set_xlabel('Month', fontsize=12)
            plt.xticks(rotation=45)
            plt.title(f'Average Monthly Bookings and Prices for {customer_type.title()} in {hotel.title()} Hotel', fontsize=14)
            plt.show()

 # This code groups the dataset by hotel, arrival_date_month, and customer_type, and calculates the average number of bookings and the
 # average price per night for each group. It then creates separate bar charts to visualize the results for each hotel and customer_type
 # group.           

 # Remarks of the analysis -> The analysis shows the average monthly number of bookings and average price per night for each combination 
 # of hotel and customer type. The charts demonstrate that there are differences in booking patterns and pricing across different types of 
 # customers and hotels. The analysis can be useful for the hotel management to understand the booking and pricing trends for different 
 # customer segments and adjust their strategies accordingly.
```

# 2) Optimal length of stay to get best daily rate?

```
# Filter the dataset to include only non-canceled bookings
df = df[df['is_canceled'] == 0]

# Calculate the daily rate for each booking
df['daily_rate'] = df['adr'] / (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])

# Group the dataset by length of stay and calculate the average daily rate for each group
grouped_data = df.groupby('stays_in_week_nights').agg(avg_daily_rate=('daily_rate', 'mean')).reset_index()

# Create a line chart to visualize the relationship between length of stay and average daily rate
plt.plot(grouped_data['stays_in_week_nights'], grouped_data['avg_daily_rate'], marker='o')
plt.xlabel('Length of Stay (Nights)', fontsize=12)
plt.ylabel('Average Daily Rate ($)', fontsize=12)
plt.title('Relationship Between Length of Stay and Average Daily Rate', fontsize=14)
plt.show()

# calculates the average daily rate for each group. It then creates a line chart to visualize the 
# relationship between length of stay and average daily rate

# Remarks of the analaysis -> The analysis shows the relationship between length of stay and the average daily rate for 
# non-canceled bookings. The data is grouped by the length of stay, and the average daily rate is calculated for each group. 
# A line chart is used to visualize the relationship between the two variables, showing that the average daily rate decreases as the 
# length of stay increases, with the steepest decline happening in the first few nights. This information can be useful for hotels in 
# setting their pricing strategy for longer stays.
```
